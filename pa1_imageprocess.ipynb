```python
#IMPORTING LIBRARIES
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
#READING 'Lenna.png' 
original_image = Image.open('Lenna.png')
plt.imshow(original_image)
```

```python
original_image.size
```

```python
#CONVERTING THE COLOURED IMAGE INTO GREY IMAGE
grayscale_image = Image.open('Lenna.png').convert('LA')
grayscale_image.save('greyscale.png')
plt.imshow(grayscale_image)
```

```python
#calculating the half downscale dimension  
half_downscale = tuple((int(grayscale_image.size[0]/2),int(grayscale_image.size[1]/2)))
half_downscale
```

```python
#Half downscale for the Nearest Neighbour method
img = grayscale_image

img_downscale_NEAREST = img.resize(half_downscale)
img_downscale_NEAREST.save('NEAREST_half_downscale.png')
plt.imshow(img_downscale_NEAREST)
```

```python
#Implementing the Nearest Neighbour Resampling Method for Upscale
img = img_downscale_NEAREST

img_upscale_NEAREST = img.resize(grayscale_image.size,resample=Image.NEAREST)
img_upscale_NEAREST.save('Neares_half_upscale.png')
plt.imshow(img_upscale_NEAREST)
```

```python
#Half downscale for the BILINEAR method
img = grayscale_image

img_downscale_BILINEAR = img.resize(half_downscale)
img_downscale_BILINEAR.save('BILINEAR_half_downscale.png')
plt.imshow(img_downscale_BILINEAR)
```

```python
#Implementing the Bilinear Resampling Method for Upscale
img = img_downscale_BILINEAR

img_upscale_BILINEAR = img.resize(grayscale_image.size,resample=Image.BILINEAR)
img_upscale_BILINEAR.save('BILINEAR_half_upscale.png')
plt.imshow(img_upscale_BILINEAR)
```

```python
#Half downscale for the BICUBIC method
img = grayscale_image

img_downscale_BICUBIC = img.resize(half_downscale)
img_downscale_BICUBIC.save('BICUBIC_half_downscale.png')
plt.imshow(img_downscale_BICUBIC)
```

```python
#Implementing the BICUBIC Resampling Method for Upscale
img = img_downscale_BICUBIC

img_upscale_BICUBIC = img.resize(grayscale_image.size,resample=Image.BICUBIC)
img_upscale_BICUBIC.save('BICUBIC_half_upscale.png')
plt.imshow(img_upscale_BICUBIC)
```

```python
#Half downscale for the lanczos method
img = grayscale_image

img_downscale_LANCZOS = img.resize(half_downscale)
img_downscale_LANCZOS.save('LANCZOS_half_downscale.png')
plt.imshow(img_downscale_LANCZOS)
```

```python
#Implementing the LANCZOS Resampling Method for Upscale
img = img_downscale_LANCZOS

img_upscale_LANCZOS = img.resize(grayscale_image.size,resample=Image.LANCZOS)
img_upscale_LANCZOS.save('LANCZOS_half_upscale.png')
plt.imshow(img_upscale_LANCZOS)
```

```python
# definning the PSNR function
def PSNR(img1,img2):
    R2 = np.amax(img1)**2
    MSE = np.sum(np.power(np.subtract(img1,img2),2))
    MSE /= (img1.size[0]*img1.size[1])
    Psnr = 10*np.log10(R2/MSE)
    return Psnr
```

```python
#Plotting the psnr values of the three resampling meethods
fig = plt.figure(figsize=(15,10))
rows = 1
columns = 3

fig.add_subplot(rows,columns,1)
plt.imshow(img_upscale_NEAREST)
plt.ylabel("Nearest Neighbour")
plt.xlabel(PSNR(img_upscale_NEAREST,grayscale_image))

fig.add_subplot(rows,columns,2)
plt.imshow(img_upscale_BILINEAR)
plt.ylabel("Bilinear")
plt.xlabel(PSNR(img_upscale_BILINEAR,grayscale_image))

fig.add_subplot(rows,columns,3)
plt.imshow(img_upscale_BICUBIC)
plt.ylabel("Bicubic")
plt.xlabel(PSNR(img_upscale_BICUBIC,grayscale_image))
```

```python
#plotting the psnr values of LANCZOS
fig.add_subplot(rows,columns,1)
plt.imshow(img_upscale_LANCZOS)
plt.ylabel("LANCZOS")
plt.xlabel(PSNR(img_upscale_LANCZOS,grayscale_image))
```

```python
img1 = cv2.imread('Lenna.png')
# displaying the image using the imshow function of matplotlib.pyplot
# first converting the image from bgr format to rgb 
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
```

```python
def ChSwap(img):
    # using the split function of cv2 
    # it splits the image into blue, green and red channels 
    b,g,r = cv2.split(img)
    rOut = b
    gOut = r
    bOut = g
    # returning the new merged image with swapped rgb channels 
    return cv2.merge([bOut,gOut,rOut])
```

```python
swap = ChSwap(img1)
# displaying the image with swapped channels
plt.imshow(cv2.cvtColor(swap, cv2.COLOR_BGR2RGB))
```

```python
fig = plt.figure(figsize=(15,10))
rows = 1
columns = 2

fig.add_subplot(rows,columns,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.ylabel("Coloured Image")

fig.add_subplot(rows,columns,2)
plt.imshow(cv2.cvtColor(swap, cv2.COLOR_BGR2RGB))
plt.ylabel("Colour band swapped version")
```

```python
#FUNCTION DEFINITIN FOR RED EYE REMOVAL

def redEyeRemoval(img):

    #image.shape returns an array having width as the first
    #element, height as 2nd as number of color channels as 3rd
    shape=img.shape

    #loop over every pixel of the image and check if it falls
    #under the category of red

    #first loop for looping over every pixel row in the image
    for i in range(shape[0]):
        #inner loop for looping over every pixel column
        for j in range(shape[1]):
            #now we will decide the criteria for a pixel to be red
            #in the numpy array of the image the pixel color values
            #are in the order b,g,r(blue,green,red)

            #below condition implies that if the blue content in the
            #pixel is in the range(0,220) and green in range(0,100) and
            #red in range(125,255) then we will change the color of that
            #pixel to red

            #img[i][j] will have an array with 3 values [b,g,r]
            #we have to alter these values for changing the color
            if(img[i][j][0]>=0 and img[i][j][1]>=0 and img[i][j][2]>=150
            and img[i][j][0]<=220 and img[i][j][1]<=100 and img[i][j][2]<=255):
                #change color to black
                img[i][j]=[0,0,0]

    #save the edited image as new image2.jpg
    cv2.imwrite("image2.jpg", img)
    #display the modified image
    #cv2.imshow("Image without red eye",img)

    img = Image.open('image2.jpg')
    plt.imshow(img)
    plt.ylabel("Image without red eye")
```

```python
#Displaying the 'redeyebaby.jpg' 
img = Image.open('redeyebaby.jpg')
plt.imshow(img)
plt.ylabel("Image with red eye")
```

```python
#reading the image as numpy array it will be a 3D array

img = cv2.imread('redeyebaby.jpg')
redEyeRemoval(img)
```
